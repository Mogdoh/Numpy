```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure() # 그래프를 그리기 위한 틀을 만든다
plt.axes() # 그안에 좌표평면들을 그린다
plt.show() # 화면에 표시
```

# 단일 그래프

### plot

    - line plot을 그려주는 함수
    - figure, axes가 없다면 이를 만들어서 그려준다

```python
np.arange(2,7)
```

```python
plt.plot(np.arange(2,7))
plt.show()
```

### 다중 그래프

    - 1개의 axes 안에 여러개의 그래프를 그린다

```python
#틀 크기(너비, 높이)

plt.figure(figsize= [15,5])

# 그래프 2개
plt.plot(np.arange(5))
plt.plot(np.arange(2,7))

plt.show()
```

### subplots

    - 한개에 여러개의 그래프를 그리기



```python
# fig는 외부의 공간 ax 그래프 공간으로 이해하면 된다
fig, ax = plt.subplots(2,2,figsize = [15,5])
print(ax.shape) 

#numpy 인덱싱과 같다
ax[0,0].plot(np.arange(5))
ax[0,1].plot(np.arange(5))
ax[0,1].plot(np.arange(2,6))
ax[1,0].plot(np.arange(5))
ax[1,1].plot(np.arange(5))
```

```python
arr = np.arange(5)
arr1 = np.arange(2,6)
plt.figure(figsize=[15,5])

plt.subplot(2,2,1)
plt.plot(arr)

plt.subplot(2,2,2)
plt.plot(arr)
plt.plot(arr1)

plt.subplot(2,2,3)
plt.plot(arr)

plt.subplot(2,2,4)
plt.plot(arr)

plt.show()
```

# 타이틀

### figure의 제목: suptitle()

```python
fig, ax = plt.subplots(1,2,figsize =(15,5))
fig.suptitle("figure title")

x = range(0,10) #
y = np.exp(x) #
ax[0].plot(x,y)

x = range(1,1000)
y = np.log(x)
ax[1].plot(x,y)

plt.show()
```

# axes의 제목: set_title()

```python
fig , ax = plt.subplots(1,2,figsize=(15,5))
fig.suptitle("figure title")

x = range(0,10) # x축
y = np.exp(x) # y축
ax[0].plot(x,y)
ax[0].set_title("exp")

x = range(1,1000) # x 축
y = np.log(x) # y 축
ax[1].plot(x,y)
ax[1].set_title("log")

plt.show()
```

# 축 제목

```python
fig , ax = plt.subplots(1,2,figsize=(15,5))
fig.suptitle("figure title")

x = range(0,10) # x축
y = np.exp(x) # y축
ax[0].plot(x,y)
ax[0].set_title("exp")
ax[0].set_xlabel("X axis")
ax[0].set_ylabel("Y axis",rotation=90,labelpad=30)

x = range(1,1000) # x 축
y = np.log(x) # y 축
ax[1].plot(x,y)
ax[1].set_title("log")
ax[1].set_xlabel("X axis")
ax[1].set_ylabel("Y axis",rotation=0,labelpad=30)

plt.show()
```

# 눈금

```python
x = range(1,1000)
y = np.log(x)
```

### 눈금회전

    - axes의 tick_params()

```python
fig,ax = plt.subplots()
ax.plot(x,y)

#눈금회전
ax.tick_params(axis="x", labelrotation = 90) #x축의 라벨이 90도 돌아간것을 볼 수 있다
plt.show()
```

### 눈금 지정

    - axes의 set_xticks(), set_yticks()

```python
fig,ax = plt.subplots(figsize= [10,10])
ax.plot(range(50))

#눈금지정
ax.set_xticks(range(50))
ax.set_yticks(range(50))

plt.show()
```

# 그래프 스타일

### marker

```python
y = np.exp(range(0,10))
fig , ax = plt.subplots(2,2,figsize=[15,10])

# marker
ax[0,0].plot(y,marker="o") #마커의 모양이 다르다
ax[0,1].plot(y,marker="v")
ax[1,0].plot(y,marker="x") 
ax[1,1].plot(y,marker="^")

plt.show()
```

### line style



```python
y = np.exp(range(0,10))
fig , ax = plt.subplots(2,2,figsize=[15,10])

# line style
ax[0,0].plot(y,marker="o")
ax[0,1].plot(y,marker="v",linestyle='--')
ax[1,0].plot(y,marker="x",linestyle='') 
ax[1,1].plot(y,marker="^")

plt.show()
```

### color

```python
y = np.exp(range(0,10))
fig , ax = plt.subplots(2,2,figsize=[15,10])

# color
ax[0,0].plot(y,marker="o")
ax[0,1].plot(y,marker="v",linestyle='--',color="y")
ax[1,0].plot(y,marker="x",linestyle='',color="r") 
ax[1,1].plot(y,marker="^",color="g")

plt.show()
```

# 범례 표시

    - axes의 legend()

```python
fig,ax = plt.subplots()
x = np.arange(10)

ax.plot(x)
ax.plot(x**2,alpha = 0.2) # alpha 투명도 0~1 사이값을 넣어주면된다.
# 범례 표시
ax.legend(["X","X^2"],loc=[0.5,0.8])

plt.show()
```

# matplotlib 심화



### 선 그래프

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.show()
```

### 수평선 그리기

    > axhline(y, xmin, xmax)
    - min과 max는 0~1 사이의 값을 입력하며 0은 왼쪽 끝 1은 오른쪽 끝

    > hlines(y, xmin, xmax)
    - 점(xmin,y)에서 점(xmax,y)를 따라 수평선을 표시함

```python
plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)

plt.show()
```

```python
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

### 수직선 그리기
    - axvline(x, ymin, ymax)
    - vlines(x, ymin, ymax)

```python
plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

### 막대 그래프
